```python
import numpy as np
import pandas as pd
import tensorflow as tf
import math
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error

import time
import datetime
now = datetime.datetime.now()
print ("Current date and time : ")
print (now.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
def metrics_time_series(y_true, y_pred):
    mae = round         (mean_absolute_error(y_true, y_pred), 7)
    mse = round         (mean_squared_error(y_true, y_pred), 7)
    rmse= round         (np.sqrt(mean_squared_error(y_true, y_pred)), 7)
    mape= round         (mean_absolute_percentage_error(y_true, y_pred), 7)
    r2  = round         (r2_score(y_true, y_pred), 7)
   
    return mae, mse, rmse, mape, r2 
```

```python
df=pd.read_excel('WBG_MAC.xlsx')
d = 'MAC'
print('Number of rows and columns:', df.shape)
df.head(5)
```

```python
do = df.DO_org[2:4800]
dow =df.DO_w[2:4800]
dob =df.DO_b[2:4800]
dog =df.DO_g[2:4800]

doew =df.DOE_w[2:4800]
doeb =df.DOE_b[2:4800]
doeg =df.DOE_g[2:4800]
```

```python
df_results = pd.DataFrame()
```

```python
name = "Weiss"
mae, mse, rmse, mape, r2 = metrics_time_series( do, dow)
res = pd.DataFrame([name, mae, mse, rmse, mape, r2], index= ["Name", "mae", "mse","rmse", "mape", "r2"]).T

df_results = df_results.append(res)
```

```python
name = "Benson"
mae, mse, rmse, mape, r2 = metrics_time_series( do, dob)
res = pd.DataFrame([name, mae, mse, rmse, mape, r2], index= ["Name", "mae", "mse","rmse", "mape", "r2"]).T

df_results = df_results.append(res)
```

```python
name = "Garcia"
mae, mse, rmse, mape, r2 = metrics_time_series( do, dog)
res = pd.DataFrame([name, mae, mse, rmse, mape, r2], index= ["Name", "mae", "mse","rmse", "mape", "r2"]).T

df_results = df_results.append(res)
```

```python
fig = plt.figure(figsize=(10,5))

plt.plot(do,  linewidth=1.0)
plt.plot(dow, linewidth=1.0)
plt.plot(dob, linewidth=1.0)
plt.plot(dog, linewidth=1.0)
#plt.plot(error, linewidth=1.0)

plt.xlabel('Days', fontsize=12)
plt.ylabel('DO (mg/L)', fontsize=12)
plt.legend(['Actual value', 'Weiss Equation','Benson & Krause Equation','Garcia & Gordon Equation'],fontsize=12, loc='best')
plt.grid(linestyle=':')#, linewidth=1)
fig.savefig('figVS1' + d + '.pdf', dpi=600)
```
